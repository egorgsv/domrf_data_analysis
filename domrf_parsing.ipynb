```python
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver
import pandas as pd
import numpy as np
import time
import requests
import json
```

```python
region=78

apiURL = 'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/object/'

browser = webdriver.Chrome(ChromeDriverManager().install())
browser.set_window_size(1920, 1280)
```

```python
url = 'https://наш.дом.рф/сервисы/каталог-новостроек/список-объектов/список?place=%d&limit=100&sortDirection=asc' %region
browser.get(url)
time.sleep(1)
pagination = browser.find_element_by_xpath('//*[@id="preloader"]/div[1]/div[3]/div[2]/div/div/div/div[2]/div[2]/div[2]/ul/li[8]/a')
try:
    pages = int(pagination.get_attribute('aria-label').split(' ')[1])
except:
    pages = 1
```

```python
houses = []
for page in range(0,pages):
    try:
        url = 'https://наш.дом.рф/сервисы/каталог-новостроек/список-объектов/список?page=%d&limit=100&sortDirection=asc&place=%d' %(page,region)
        browser.get(url)
        
        elements = browser.find_elements_by_xpath('//*[@id="preloader"]/div[1]/div[3]/div[1]/div/div/div[2]/*')
        for row in elements:
            try:
                
                estate = {}
                
                estate_id = row.get_attribute("href").split("/")[-1]
                estate['id'] = estate_id
                
                url_json = apiURL + estate_id
                estate_json = ''
                try:
                    response = requests.get(url_json)
                    estate_json = json.loads(response.text)
                except:
                    print("responded with non-200 code for object", estate_id, ": ", response)
                estate['region'] = estate_json['data']['region']
                estate['address'] = estate_json['data']['address']
                estate['developer'] = estate_json['data']["developer"]["devFullCleanNm"]
                estate['lat'] = estate_json['data']['objLkLatitude']
                estate['lon'] = estate_json['data']['objLkLongitude']
                estate['class_desc'] = estate_json['data']['objLkClassDesc']
                try:
                    estate['plan_ready100perc_date'] = estate_json['data']['objReady100PercDt']
                except:
                    print("no data")
                try:
                    estate['floors'] = estate_json['data']['floorMax']
                except:
                    print("no data")
                try:
                    estate['flats'] = estate_json['data']['objElemLivingCnt']
                except:
                    print("no data")
                try:
                    estate['square'] = estate_json['data']['objSquareLiving']
                except:
                    print("no data")
                try:
                    estate['object_problem'] = estate_json['data']['objProblemFlg']
                except:
                    print("no data")
                try:
                    estate['wall_material'] = estate_json['data']['wallMaterialShortDesc']
                except:
                    print("no data")
                try:
                    estate['ceiling_height'] = estate_json['data']['objLivCeilingHeight']
                except:
                    print("no data")    
                try:
                    estate['price_avg'] = estate_json['data']['objPriceAvg']
                except:
                    print("no data")
                try:
                    estate['transport_dist_index'] = estate_json['data']['transportDistIndexValue']
                except:
                    print("no data")
                try:
                    estate['infrastructure_index'] = estate_json['data']['infrastructureIndexValue']
                except:
                    print("no data")
                try:
                    estate['parking_count'] = estate_json['data']['objElemParkingCnt']
                except:
                    print("no data")
                try:
                    estate['guest_parking_count'] = estate_json['data']['objInfrstrObjPrkngCnt']
                except:
                    print("no data")   
                try:
                    estate['energy_eff'] = estate_json['data']['objEnergyEffShortDesc']
                except:
                    print("no data")    
                    
                
                print(estate)
                houses.append(estate)
            except Exception as e:
                print('smth went wrong: %s' %e)
            pass
    except Exception as e:
        print('smth critical went wrong: %s' %e)
    
```

```python
houses = pd.DataFrame(houses)
```

```python
houses.to_csv('futureEstate_%dregion.csv' % region, index = False)
```
